```
#generate the shape dict and diacritics dict, which are used in the Arabic word class
#to determine the shapes and sounds of Arabic letters in any given word
shape_dict={} #the key is the letter ID
ar_shape_dict={} #the key is the letter (ar)
name_dict={}
for row in full_sheet.iterrows():
  cur_obj={}
  row_dict=row[1].to_dict()
  key=row_dict["arabic_script_extended"]
  letter_ar=row_dict["arabic_script"]
  shapes=[v.strip() for v in row_dict["letter_shapes"].split() if v.strip()]
  shape_dict[key]=shapes
  ar_shape_dict[letter_ar]=shapes
  #name_dict[key]=row_dict["letter_name"]
  #print(row_dict)


#Get the diacritics
diacritics_dict={}
diacritics_list=[]
for row in diacritics_sheet.iterrows():
  row_dict=row[1].to_dict()
  key=row_dict["name"]
  item=row_dict["item"]
  diacritics_dict[key]=item
  diacritics_list.append(item)
  #print(row_dict)

#Converting words to latin characters
romanization_dict={}
for row in conversion_sheet.iterrows():
  row_dict=row[1].to_dict()
  key=row_dict["ar"].strip()
  vals=str(row_dict["en"]),str(row_dict["en_beginning"])
  romanization_dict[key]=vals
print(romanization_dict)

ar_shape_dict={'أ': ['أ', 'ـأ'], 'ب': ['ب', 'بـ', 'ـبـ', 'ـب'], 'ت': ['ت', 'تـ', 'ـتـ', 'ـت'], 'ث': ['ث', 'ثـ', 'ـثـ', 'ـث'], 'ج': ['ج', 'جـ', 'ـجـ', 'ـج'], 'ح': ['ح', 'حـ', 'ـحـ', 'ـح'], 'خ': ['خ', 'خـ', 'ـخـ', 'ـخ'], 'د': ['د', 'ـد'], 'ذ': ['ذ', 'ـذ'], 'ر': ['ر', 'ـر'], 'ز': ['ز', 'ـز'], 'س': ['س', 'سـ', 'ـسـ', 'ـس'], 'ش': ['ش', 'شـ', 'ـشـ', 'ـش'], 'ص': ['ص', 'صـ', 'ـصـ', 'ـص'], 'ض': ['ض', 'ضـ', 'ـضـ', 'ـض'], 'ط': ['ط', 'طـ', 'ـطـ', 'ـط'], 'ظ': ['ظ', 'ظـ', 'ـظـ', 'ـظ'], 'ع': ['ع', 'عـ', 'ـعـ', 'ـع'], 'غ': ['غ', 'غـ', 'ـغـ', 'ـغ'], 'ف': ['ف', 'فـ', 'ـفـ', 'ـف'], 'ق': ['ق', 'قـ', 'ـقـ', 'ـق'], 'ك': ['ك', 'كـ', 'ـكـ', 'ـك'], 'ل': ['ل', 'لـ', 'ـلـ', 'ـل'], 'م': ['م', 'مـ', 'ـمـ', 'ـم'], 'ن': ['ن', 'نـ', 'ـنـ', 'ـن'], 'هـ': ['ه', 'هـ', 'ـهـ', 'ـه'], 'و': ['و', 'ـو'], 'ي': ['ي', 'يـ', 'ـيـ', 'ـي'], 'ء': ['ء'], 'لا': ['لا', 'ـلا'], 'ئ': ['ئ', 'ئـ', 'ـئـ', 'ـئ'], 'ؤ': ['ؤ', 'ـؤ'], 'ة': ['ة', 'ـة'], 'ى': ['ى', 'ـى'], 'إ': ['إ', 'ـإ'], 'آ': ['آ', 'ـآ'], 'لأ': ['لأ', 'ـلأ'], 'لإ': ['لإ', 'ـلإ'], 'لآ': ['لآ', 'ـلآ']}    
romanization_dict={'أ': ('2', 'a'), 'ا': ('aa', 'a'), 'إ': ('2i', 'i'), 'آ': ('2aa', 'aa'), 'ى': ('aa', ''), 'ب': ('b', ''), 'ت': ('t', ''), 'ث': ('th', ''), 'ج': ('j', ''), 'ح': ('7', ''), 'خ': ('kh', ''), 'د': ('d', ''), 'ذ': ('dh', ''), 'ر': ('r', ''), 'ز': ('z', ''), 'س': ('s', ''), 'ش': ('sh', ''), 'ص': ('S', ''), 'ض': ('D', ''), 'ط': ('T', ''), 'ظ': ('DH', ''), 'ع': ('3', ''), 'غ': ('gh', ''), 'ف': ('f', ''), 'ق': ('q', ''), 'ك': ('k', ''), 'ل': ('l', ''), 'م': ('m', ''), 'ن': ('n', ''), 'ه': ('h', ''), 'و': ('oo', 'w'), 'ي': ('ee', 'y'), 'ة': ('h', ''), 'ئ': ('2', ''), 'ؤ': ('2', ''), 'ء': ('2', ''), 'َ': ('a', ''), 'ِ': ('i', ''), 'ُ': ('u', ''), 'ٌ': ('un', ''), 'ً': ('an', ''), 'ٍ': ('in', ''), 'ّ': ('x', '')}

class word_class: #analyze a word with diacritics in different ways
  def __init__(self,word0,shape_dict0,romanize_dict0):
    self.word=word0
    self.shape_dict=shape_dict0
    self.romanize_dict=romanize_dict0
    #first split the word into chunks, based on diacritics
    self.chunks=[]
    new_chunk=""
    self.plain=""
    for w0 in self.word:
      if w0.isalpha() or w0==" ":
        self.plain+=w0
        if  new_chunk!="": self.chunks.append(new_chunk)
        new_chunk=w0 
      else: new_chunk+=w0
    if  new_chunk!="": self.chunks.append(new_chunk)
    #get the shape
    self.word_letter_shapes_plain=[]
    self.word_letter_shapes=[]
    prev_shapes=[]
    
    for i0,ch in enumerate(self.chunks):
      att_before,att_after=False,False #attach before and attach after
      w0=ch[0]
      cur_shape=w0
      possible_shapes=shape_dict0.get(w0,[])

      if i0==0 and len(possible_shapes)>2: att_after=True #cur_shape=w0+"ـ"
      elif i0==len(self.chunks)-1:
        if len(prev_shapes)>2: att_before=True #cur_shape="ـ"+w0
      else:
        if len(prev_shapes)>2: att_before=True # cur_shape="ـ"+w0
        if len(possible_shapes)>2:  att_after=True
          #cur_shape=cur_shape+"ـ"
      prev_shapes=possible_shapes
      cur_shape=w0
      if att_before: cur_shape="ـ"+cur_shape
      if att_after: cur_shape=cur_shape+"ـ"
      self.word_letter_shapes_plain.append(cur_shape)
      #print(cur_shape)
      
      cur_shape=ch
      if att_before: cur_shape="ـ"+cur_shape
      if att_after: cur_shape=cur_shape+"ـ"
      self.word_letter_shapes.append(cur_shape)
  
    #get the sound
    self.romanized_chunks=[]
    prev_romanized=""
    for ch_i,ch in enumerate(self.chunks):
      found=self.romanize_dict.get(ch[0],ch[0])
      first=found[0]
      if ch_i==0 and found[1]!="": first=found[1]
      #first=self.romanize_dict[ch[0]][0]
      
      #cur_romanized_chunk=first
      dia0="".join([self.romanize_dict.get(v,["",""])[0] for v in ch[1:]]) #convert the diacritics part of the chunk
      has_shaddah=False
      if "x" in dia0: has_shaddah=True #in the conversion sheet, shadda is converted to x
      dia0=dia0.replace("x","")
      if dia0 and found[1]!="": first=found[1] 
      if ch_i>0 and "2" in found[0]: first="2" #imra2ah

      if prev_romanized.endswith("a") and ch[0] in "وي" and found[1]!="": first=found[1]
      if has_shaddah and found[1]!="": first=found[1]

      cur_romanized_chunk=first+dia0
      if has_shaddah: cur_romanized_chunk=first+cur_romanized_chunk #shaddah

      if ch_i==0 and cur_romanized_chunk=="aa" and ch[0]!="آ":  cur_romanized_chunk="a" #cheap hacks to fix certain fringe cases
      if cur_romanized_chunk=="ai":  cur_romanized_chunk="i"
      if cur_romanized_chunk=="ii":  cur_romanized_chunk="i"
      if cur_romanized_chunk=="au":  cur_romanized_chunk="u"
      if cur_romanized_chunk=="aan":  cur_romanized_chunk="an"




      self.romanized_chunks.append(cur_romanized_chunk)
      #print(cur_romanized_chunk)
      prev_romanized=cur_romanized_chunk
    self.romanized=""
    for s_i, rom in enumerate(self.romanized_chunks):
      next_chunk=""
      if s_i<len(self.romanized_chunks)-1: next_chunk=self.romanized_chunks[s_i+1]
      if rom.endswith("i") and next_chunk=="ee": rom=rom[:-1] #avoid iee for kasrah before yaa2
      if rom.endswith("u") and next_chunk=="oo": rom=rom[:-1]
      if rom.endswith("a") and next_chunk=="aa": rom=rom[:-1]
      self.romanized+=rom
    



word_dia="وَسَنُشَدِّد"
word_dia="مُشَدَّدَة"
word_dia="يَسْتَبِدُّ"
# word_dia="وَكَإنَّها"
# word_dia="سَيَكون"
word_dia="تُفَّاحَة"
# word_dia="وَسَائِل"
# word_dia="مَصائب"
# word_dia="يَذهَبون"
# word_dia="قِراءَة"
# word_dia="يُساعِدونَهُم"
# word_dia="بِبَساطَة"
# word_dia="نَقِيّ"



#word_dia="شُكراً"
#word_dia="يَومٌ"
#word_dia="لَيسَ"
#word_dia="بؤس"

arw_obj=word_class(word_dia, ar_shape_dict, romanization_dict)
print(arw_obj.chunks)
print(arw_obj.romanized_chunks)
print(arw_obj.word_letter_shapes)
print(arw_obj.word_letter_shapes_plain)



# print(ar_shape_dict)
# print(romanization_dict)
# print(en2ar_name_dict)
# print(ar2en_name_dict)
# print(all_letters_names_romanized)
```

```
#Here we create the main json data file for the game
#Get the names and shapes of letters
import json
from itertools import groupby
json_fname="arabic_data.json"
json_fopen=open(json_fname,"w")
json_dict={}
#get the test data
test_questions=[]
for row in test_sheet.iterrows():
  cur_obj=row[1].to_dict()
  cur_obj["options"]=[cur_obj["correct_answer"],cur_obj["option1"],cur_obj["option2"],cur_obj["option3"]]
  test_questions.append(cur_obj)
json_dict["test_questions"]=test_questions
#Get the main levels
main_levels=[]
for row in main_levels_sheet.iterrows(): main_levels.append(row[1].to_dict())
json_dict["main_levels"]=main_levels
#Get the alphabet sublevels
alphabet_sublevels=[]
for row in alphabet_sublevels_sheet.iterrows(): alphabet_sublevels.append(row[1].to_dict())
json_dict["alphabet_sublevels"]=alphabet_sublevels
#Get the sounds sublevels
sounds_sublevels=[]
for row in sounds_sublevels_sheet.iterrows(): sounds_sublevels.append(row[1].to_dict())
json_dict["sounds_sublevels"]=sounds_sublevels
#Get the diacritics sublevels
diacritics_sublevels=[]
for row in diacritics_sublevels_sheet.iterrows(): diacritics_sublevels.append(row[1].to_dict())
json_dict["diacritics_sublevels"]=diacritics_sublevels

#Now the alphabet lessons
alphabet_lesson_list=[]
for row in alphabet_lessons_sheet.iterrows():alphabet_lesson_list.append(row[1].to_dict())
alphabet_lessons_grouped=[(key,list(group)) for key,group in groupby(alphabet_lesson_list,lambda x:x["level"])]
json_dict["alphabet_lessons"]=dict(iter(alphabet_lessons_grouped)) 

#Now the sound lessons
sound_lesson_list=[]
for row in sounds_lessons_sheet.iterrows():sound_lesson_list.append(row[1].to_dict())
sound_lessons_grouped=[(key,list(group)) for key,group in groupby(sound_lesson_list,lambda x:x["level"])]
json_dict["sound_lessons"]=dict(iter(sound_lessons_grouped))

#Now the diacrtitics lessons
diacritics_lesson_list=[]
for row in diacritics_lessons_sheet.iterrows():diacritics_lesson_list.append(row[1].to_dict())
diacritics_lessons_grouped=[(key,list(group)) for key,group in groupby(diacritics_lesson_list,lambda x:x["level"])]
json_dict["diacritics_lessons"]=dict(iter(diacritics_lessons_grouped))

#Now the keyboard layout
keyboard_dict={}
reverse_keyboard_dict={}

for row in keyboard_sheet.iterrows():
  row_dict=row[1].to_dict()
  row_id=row_dict["row"]
  primary=row_dict["primary"]
  secondary=row_dict["secondary"]  
  keyboard_dict[row_id]=(primary,secondary)

  reverse_keyboard_dict[primary]=(row_id,)
  if secondary: reverse_keyboard_dict[secondary]=(row_id,"shift")

json_dict["keyboard_dict"]=keyboard_dict
json_dict["reverse_keyboard_dict"]=reverse_keyboard_dict


#Now get the shapes and names for letters
all_ids=[] #for letters and other characters
letter_ids=[] #only for letters
letter_dict={}
shape_dict={}
name_dict={}
for row in alphabet_lessons_sheet.iterrows():
  cur_obj={}
  row_dict=row[1].to_dict()
  row_dict_temp={}
  letter_id=row_dict["id"]
  all_ids.append(letter_id)
  name_dict[letter_id]=row_dict["letter_name"]
  cur_obj["letter_name"]=row_dict["letter_name"]
  cur_obj["arabic_letter_name"]=row_dict["arabic_letter_name"]
  cur_obj["romanization"]=row_dict["romanization"]
  arabic_script=row_dict["arabic_script"]
  #print(letter_id,arabic_script)
  arabic_script=row_dict["arabic_script"]
  if arabic_script!="": letter_ids.append(letter_id)
  
  # if str(arabic_script).lower()=="":  arabic_script=None
  # if arabic_script!=None: letter_ids.append(letter_id)
  cur_obj["arabic_script"]=arabic_script

  #cur_obj["arabic_script_extended"]=row_dict["arabic_script_extended"]
  cur_shapes=[v.strip() for v in row_dict["letter_shapes"].split() if v.strip()]
  cur_obj["shapes"]=cur_shapes
  letter_dict[letter_id]=cur_obj
  shape_dict[letter_id]=cur_shapes
  #print(letter_id,cur_obj)

json_dict["all_ids"]=all_ids
json_dict["letter_ids"]=letter_ids
json_dict["letter_dict"]=letter_dict
json_dict["shape_dict"]=shape_dict
json_dict["name_dict"]=name_dict

#print(letter_dict)
#Now let's get the list of sounds
sound_ids=[]
sound_dict={}
for row in sounds_lessons_sheet.iterrows():
  cur_obj={}
  row_dict=row[1].to_dict()
  sound_id=row_dict["id"]
  sound_ids.append(sound_id)
  cur_obj["sound"]=row_dict["sound"]
  cur_obj["arabic"]=row_dict["letter_diacritic"]
  sound_dict[sound_id]=cur_obj
json_dict["sound_ids"]=sound_ids
json_dict["sound_dict"]=sound_dict
#we may add a list of sounds with their corresponding diacritics to form questions for the diacritics levels

#and now the list of words, and which sounds and letters they have
all_words=[]
word_dict={}
vocabulary_words=[]
animal_words=[]

for row in vocabulary_sheet.iterrows():
  row_dict=row[1].to_dict()
  ar_word=row_dict["arabic"]
  en_word=row_dict["english"]
  vocabulary_words.append((ar_word,en_word))
  all_words.append((ar_word,en_word))
for row in animals_sheet.iterrows():
  row_dict=row[1].to_dict()
  ar_word=row_dict["arabic"]
  en_word=row_dict["english"]
  animal_words.append((ar_word,en_word))
  all_words.append((ar_word,en_word))

word_dict={}
for ar_word,en_word in all_words:

  if " " in ar_word: continue
  arw_obj1=word_class(ar_word, ar_shape_dict, romanization_dict)
  
  local_word_dict={}
  local_word_dict["arabic"]=ar_word
  local_word_dict["english"]=en_word
  local_word_dict["chunks"]=arw_obj1.chunks
  local_word_dict["plain"]=arw_obj1.plain
  local_word_dict["letters"]=list(arw_obj1.plain)
  local_word_dict["romanized"]=arw_obj1.romanized
  local_word_dict["romanized_chunks"]=arw_obj1.romanized_chunks
  local_word_dict["word_letter_shapes"]=arw_obj1.word_letter_shapes
  local_word_dict["word_letter_shapes_plain"]=arw_obj1.word_letter_shapes_plain
  word_dict[arw_obj1.romanized]=local_word_dict
  #print(ar_word)
  #print(local_word_dict)
  #print("-----")  

json_dict["word_dict"]=word_dict

#now we get the romanization of animal names --> animal IDs 
animals_ids=[]
for ar,en in animal_words:
  if " " in ar: continue
  arw_obj1=word_class(ar, ar_shape_dict, romanization_dict)  
  animals_ids.append(arw_obj1.romanized)
json_dict["animals_ids"]=animals_ids
print(animals_ids)

#and we get the similar words to each animal name 
import difflib
similar_dict={}
for a1 in animals_ids:
  corr_words=[]
  for a2 in animals_ids:
    if a1==a2: continue
    similarity = difflib.SequenceMatcher(None, a1, a2).ratio()
    corr_words.append((a2,similarity))
  corr_words.sort(key=lambda x:-x[-1])
  similar_dict[a1]=[v[0] for v in corr_words[:6]]
  # print(a1)
  # print(corr_words[:10])
  # print("-----")
json_dict["similar_dict"]=similar_dict

#print(letter_ids)
json_dict["fat7ah"]="َ"
json_dict["kasrah"]="ِ"
json_dict["dammah"]="ُ"
json_dict["shaddah"]="ّ"



json_str=json.dumps(json_dict)
#json_str=json_str.replace("NaN","null")

json_fopen.write(json_str)
json_fopen.close() 
```

```
#split each Arabic word to its individual letters with the corresponding shapes
word="وسوف"

def split_ar_word(word,cur_shape_dict):
  word_letter_shapes=[]
  prev_shapes=[]
  for i0,w0 in enumerate(word):
    cur_shape=w0
    possible_shapes=cur_shape_dict.get(w0,[])

    if i0==0 and len(possible_shapes)>2:cur_shape=w0+"ـ"
    elif i0==len(word)-1 or word[i0+1]==" ":
      if len(prev_shapes)>2: cur_shape="ـ"+w0
    else:
      if len(prev_shapes)>2: cur_shape="ـ"+w0
      if len(possible_shapes)>2: 
        cur_shape=cur_shape+"ـ"
    prev_shapes=possible_shapes
    word_letter_shapes.append(cur_shape)
  return word_letter_shapes

word1="الوحوش"
word1="والاستعمار"
#word1="للتنمية والمساعدات الإنمائية في هذا البلد الصغير"
split_ar_word(word1,shape_dict)
```
